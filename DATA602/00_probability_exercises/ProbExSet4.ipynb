### Problem 1

```r
lower = 0
upper = 1
f = function(x){6*x*(1-x)}
# A
print(round(integrate(f, lower=0.26, upper=0.48)$value, 4))
# B
print(round(integrate(f, lower=0.81, upper=1)$value, 4))
# C
# Expected value - integret x*f(x)
ef = function(x){x * f(x)}
e = integrate(ef, lower=lower, upper=upper)$value
print(round(e, 2))
# Standart deviation - sqrt(E(x^2) - E(x)^2)
m2 = function(x){x^2 * f(x)}
a = integrate(m2, lower=lower, upper=upper)$value
stdev = sqrt(a - e^2)
print(round(stdev, 2))
```
