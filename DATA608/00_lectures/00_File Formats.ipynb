# Binary Files

It is possible to use Python to read and write binary files directly to file. This is often used for other data formats such as music, images, or videos. You can also design your own binary format, if you feel you have a more efficient format to encode your data (but take care to ensure that the CPU/memory resources you need to encode and decode your data are minimal!)

## Pickling

Python has a native format for writing objects to file. The process is known as pickling. Pickling is not guaranteed to be as space-efficient as file compression, but tends to run quite quickly.

```python
import pickle

simple_dict = {'Alice': 3, 'Bob': 4, 'Carol': 9, 'Eve': 10}

with open('simple.bin', 'wb') as f3:
    pickle.dump(simple_dict, f3)
    f3.close()
    
    
with open('simple.bin', 'rb') as f4:
    another_dict = pickle.load(f4)
    print (another_dict)
    f4.close()
```

# File compression

## Zipping

It is possible to have pandas zip your data as you store them. Use the `compression` argument when you write your csv to do so. Several different compression algorithms are available, choose them to fit the requirements of your system. 

(Note that the `zip` format requires that only one data file can be in your zip file!)

Reading zipped data

Similar to writing a csv, pandas can also read directly from a compressed file. The default value of the `compression` argument is "infer," which means that pandas will try to do its best to provide an appropriate decomporession algorithm.



```python
df = pd.read_csv("Traffic_Counts_at_Permanent_stations.csv")

df.head()
df.to_csv('traffic.csv.gz', compression='gzip')

df2 = pd.read_csv('traffic.csv.gz', compression = 'gzip')
df2.head()
```
